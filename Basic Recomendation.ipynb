```python
df_new = df[['keywords','cast','genres','director']]
```

```python
df_new1 = df_new['keywords'] + df_new['cast'] + df_new['genres'] + df_new['director']
```

```python
df_new['combine_feature'] = df_new1
```

```python
df_new.dropna()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df_new.dropna()
```

```python
features = ['keywords','cast','genres','director','combine_feature']
```

```python
for feature in features:
    df_new[feature] = df_new[feature].fillna('')
```

```python
count_matrix = cv.fit_transform(df_new["combine_feature"])
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
def find_title_from_index(index):
    return df[df.index == index]["title"].values[0]
def find_index_from_title(title):
    return df[df.title == title]["index"].values[0]
```

```python
movie = "Avatar"
movie_index = find_index_from_title(movie)
```

```python
similar_movies = list(enumerate(cosine_sim[movie_index]))
```

```python
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
```

```python
i=0
for element in sorted_similar_movies:
    print(find_title_from_index(element[0]))
    i=i+1
    if i>0:
        break
```
